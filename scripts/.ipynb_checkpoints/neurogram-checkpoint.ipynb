```python
import os
import pickle
from dataclasses import dataclass, field
from typing import Union
import argparse

import numpy as np
import librosa
import scipy
import matplotlib.pyplot as plt
import matplotlib.ticker as mplticker
import soundfile as sf

import bruce

DATA = os.path.realpath("../data")
OUTPUT = os.path.join(DATA, "data")
REC = os.path.join(OUTPUT, "reconstructed")
assert os.path.isdir(REC)

BRUCE_MIN_FREQ = 125
BRUCE_MAX_FREQ = 20_000

MIN_FREQ = 150
MAX_FREQ = 10500

n_mels = 40
mel_scale = librosa.filters.mel_frequencies(n_mels, fmin=MIN_FREQ, fmax=MAX_FREQ, htk=False) 

SEED = 12
bruce.set_seed(SEED)
np.random.seed(SEED)

@dataclass
class Neurogram:
    dt: float
    sample_rate: int 
    f: np.array = field(repr=None)
    t: np.array = field(repr=None)
    data: np.array = field(repr=None)
    stim: bruce.stimulus.Stimulus = None
    name: str = None   
    
    def save(self, path: str):
        with open(path, "wb") as f:    
            pickle.dump(self, f) 

    @staticmethod
    def get(
        path: str, 
        cfs: Union[int, np.array] = 40, 
        bin_width: float = 1e-4, 
        n_rep: int = 1
    ) -> "Neurogram":
        stim = bruce.stimulus.from_file(path, True)
        
        ng = bruce.Neurogram(cfs)
        ng.bin_width = bin_width or stim.time_resolution
        ng.create(stim, n_rep)
        
        data = ng.get_unfiltered_output()
        y = ng.get_cfs()
        t = np.arange(data.shape[1]) * ng.bin_width
        return Neurogram(ng.bin_width, int(np.ceil(1 / ng.bin_width)), y, t, data, stim,  name = os.path.basename(path))

def signal_noise_ratio(preds: np.array, target: np.array, zero_mean: bool = True):
    eps = np.finfo(preds.dtype).eps
    
    if zero_mean:
        target = target - np.mean(target)
        preds = preds - np.mean(preds)
        
    noise = target - preds

    snr_value = (np.sum(target**2) + eps) / (np.sum(noise**2) + eps)
    return 10 * np.log10(snr_value)
```

```python
ng = Neurogram.get(f'{DATA}/defineit.wav', cfs=mel_scale)
```

```python
S = librosa.feature.melspectrogram(y=np.array(ng.stim.data), sr=ng.stim.sampling_rate, n_mels=n_mels, fmin=MIN_FREQ, fmax=MAX_FREQ)
# S_dB = librosa.power_to_db(S, ref=np.max)
# fig, ax = plt.subplots(figsize=(9, 4))
# img = librosa.display.specshow(S_dB, x_axis='time', y_axis='mel', sr=ng.stim.sampling_rate, fmin=MIN_FREQ, fmax=MAX_FREQ, ax=ax)
# fig.colorbar(img, ax=ax);
```

```python
def min_max_scale(data, a=-80, b=0):
    data_min = np.min(data)
    data_max = np.max(data)
    return a + (data - data_min) * (b - a) / (data_max - data_min)

def apply_filter(
    ng: Neurogram,
    window_type: str = "hann",
    window_size: int = 2048,
    hop_length: int = 512
):
    window = scipy.signal.get_window(window_type, window_size)
    wsum = window.sum()
    data = np.vstack(
        [
            (np.convolve(ng.data[i], window, mode="same") / wsum)[::hop_length]
             for i in range(ng.data.shape[0])
        ]
    )
    data = min_max_scale(data)
    return data

def plot_heatmap(t, y, data, ax = None, fig=None):
    if ax is None:
        fig, ax = plt.subplots(figsize=(9, 4))
        
    img = ax.pcolormesh(
        t, y, data, cmap="inferno", vmin=data.min(), vmax=data.max()
    )
    ax.set_yscale("symlog", linthresh=1000.0, base=2)
    ax.set_xlabel("time [s]")
    ax.yaxis.set_major_formatter(mplticker.ScalarFormatter())
    ax.yaxis.set_major_locator(mplticker.SymmetricalLogLocator(ax.yaxis.get_transform()))
    ax.yaxis.set_label_text("Hz")
    fig.colorbar(img, ax=ax)
        
window_size = 256
hop_length = window_size // 4
data = apply_filter(
    ng,     
    window_type = "hann",
    window_size = window_size,
    hop_length = hop_length
)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 4))
plot_heatmap(ng.t[::hop_length], ng.f, data, ax1, fig)
ax1.set_title("neurogram (bruce)")

S_db = librosa.power_to_db(S, ref=np.max)
plot_heatmap(np.linspace(0, ng.stim.stimulus_duration, len(S_db[0])), ng.f, S_db, ax2, fig)
ax2.set_title("spectrogram (sound)")

ref_db = np.max(S)
mel_spec_ng = librosa.db_to_power(data, ref=ref_db)
mel_spec_so = librosa.db_to_power(S_db, ref=ref_db)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 4))
plot_heatmap(ng.t[::hop_length], ng.f, mel_spec_ng, ax1, fig)
ax1.set_title("neurogram (bruce)")
plot_heatmap(np.linspace(0, ng.stim.stimulus_duration, len(S_db[0])), ng.f, mel_spec_so, ax2, fig)
ax2.set_title("spectrogram (sound)")
```

```python
y = np.array(ng.stim.data)
sr = ng.stim.sampling_rate

mel_inversed = librosa.feature.inverse.mel_to_audio(mel_spec_so, sr=sr)
ng_inversed  = librosa.feature.inverse.mel_to_audio(mel_spec_ng, sr=int(1 / 5e-4))
```

```python
ng_inversed = scipy.signal.resample(ng_inversed, len(mel_inversed))
```

```python
wlen = 32
order = 2
n_samples = 100000

plt.figure(figsize=(10, 8))

# plt.plot(mel_inversed[:n_samples], label="inverted spectrogram (sound-based)", color="blue", alpha=.8)
# plt.plot(ng_inversed[:n_samples], label="inverted neurogram (bruce-based)", color= "red", alpha=.8)

ng1 = ng_inversed.copy()
plt.plot(
    np.linspace(0, ng.stim.stimulus_duration, len(y)),
    y, label="original signal", color="black", linewidth=2)
plt.plot(
    np.linspace(0, ng.stim.stimulus_duration, len(mel_inversed)),
    scipy.signal.savgol_filter(mel_inversed, wlen, order), 
         label="inverted spectrogram (sound-based) (filtered)", color ="blue", linestyle="dashed", alpha=.8)
plt.plot(
    np.linspace(0, ng.stim.stimulus_duration, len(ng1)),
    scipy.signal.savgol_filter(ng1, wlen, order), 
         label="inverted neurogram (bruce-based) (filtered)", color="red", linestyle="dashed", alpha=.8)
plt.grid()
plt.legend()


sf.write(
     f"{REC}/mel_reconstructed_{ng.name}", mel_inversed, sr, subtype='PCM_24'
)
sf.write(
     f"{REC}/ng_reconstructed_{ng.name}", ng1, sr, subtype='PCM_24'
)

sf.write(
     f"{REC}/mel_reconstructed_filtered_{ng.name}", scipy.signal.savgol_filter(mel_inversed, wlen, order), sr, subtype='PCM_24'
)
sf.write(
     f"{REC}/ng_reconstructed_filtered_{ng.name}", scipy.signal.savgol_filter(ng1, wlen, order), sr, subtype='PCM_24'
)
```
